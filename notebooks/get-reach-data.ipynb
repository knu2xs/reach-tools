```python
import json
from pathlib import Path
import re

from tqdm.notebook import tqdm

from reach_tools.utils.procure import download_raw_json_from_aw
```

```python
dir_data = Path.cwd().parent / 'data'
dir_raw = dir_data / 'raw'
dir_raw_aw = dir_raw / 'american_whitewater'

dir_raw_aw
```

```python
if not dir_raw_aw.exists():
    dir_raw_aw.mkdir(parents=True)
```

```python
fail_count = 0
max_fail = 1000
max_range = 10000000
```

```python
for reach_pth in dir_raw_aw.glob(r'aw_*.json'):
    break

```

```python
downloaded_reach_ids = [int(re.search(r"aw_(\d+).json", reach_pth.name).group(1)) for reach_pth in dir_raw_aw.glob(r'aw_*.json')]

len(downloaded_reach_ids)
```

```python
for reach_id in range(1, max_range):

    file_pth = dir_raw_aw / f'aw_{reach_id:08d}.json'

    if reach_id in downloaded_reach_ids:
        pass

    else:

        try:
            # download the data from AW
            aw_json = download_raw_json_from_aw(reach_id)
    
            # if the file exists, clean it out
            if file_pth.exists():
                file_pth.unlink()
    
            # save to a local file
            with open(file_pth, 'x') as fp:
                json.dump(aw_json, fp, indent=2)
    
            fail_count = 0
    
            print(f'Downloaded reach_id={reach_id}')
    
        except:
            fail_count += 1
            print(f'No reach_id={reach_id} (fail_count: {fail_count})')
            
            if fail_count > max_fail:
                break
```
